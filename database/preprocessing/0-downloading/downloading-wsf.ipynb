```python
import numpy as np
import itertools
import wget
```

```python
EW = np.arange(-12, 36, 2).tolist()
NS = np.arange(32, 72, 2).tolist()
```

```python
tiles = list(itertools.product(EW, NS))
print(tiles[:5])
print(tiles[-5:])
```

```python
# start_url = 'https://download.geoservice.dlr.de/WSF2019/files/WSF2019_v1_'
start_url = r'https://download.geoservice.dlr.de/WSF_EVO/files/WSFevolution_v1_'
```

```python
links = [start_url+str(item[0])+'_'+str(item[1])+'.tif' for item in tiles]
print(links[:5])
```

```python
# example
wget.download(links,out='E:\data-eubucco\wfs-evo')
```

```python
# full loop
for n,link in enumerate(links):
    try:
        wget.download(link,'E:\data-eubucco\wfs-evo')
    except:
        print(f'Tile {n} does not exist.')
```
