# Get a list of all teryt IDs 
We  get a list of all TERYT IDs from here: https://eteryt.stat.gov.pl/eTeryt/rejestr_teryt/udostepnianie_danych/baza_teryt/uzytkownicy_indywidualni/pobieranie/pliki_pelne.aspx

We then filter for 'powiat’ in NAZWA_DOD column to get TERYT IDs on county level. Building up on those ID's we can construct the download links by combining columns ‘WOJ’ and ‘POW’. In example below ‘boleslawiecki’ county is WOJ: ‘02’ + POW: ‘01’ so the link will be
https://integracja.gugik.gov.pl/Budynki3D/pobierz.php?d=2&plik=powiaty/lod1/0201_gml.zip

```python
# read in csv
df = pd.read_csv('.../teryt_list.csv', encoding = "ISO-8859-1")

# rename columns with strange characters
df = df.rename(columns={'ï»¿WOJ':'woj','POW':'pow'})

# convert to str
df['woj'] = df['woj'].astype(str)
df['pow'] = df['pow'].astype(str)

# add zeros where we only have one digit
for idx, val in enumerate(df.woj):
    #print(idx)
    if len(val)<2:
        df.woj[idx]='0'+val

for idx2, val2 in enumerate(df['pow']):
    if len(val2)<2:
        df['pow'][idx2]='0'+val2

# add str of both columns to receive teryt ID
df['teryt'] = df['woj']+df['pow']

# show df
df
```

# Let's download!

```python
from bs4 import BeautifulSoup
from bs4.element import Comment
import pandas as pd
import numpy as np
import requests

import requests
from lxml import etree
import urllib.parse
import urllib.request
```

```python
# Path of output dir
path_download_dir = "/Volumes/ssd_500gb/Drive/Downloads/poland/"
url1 = 'https://integracja.gugik.gov.pl/Budynki3D/pobierz.php?d=2&plik=powiaty/lod1/'
url2 = '_gml.zip'
```

```python
for val in df.teryt:
    path = path_download_dir+val+'_gml.zip'
    url = url1+val+url2
    urllib.request.urlretrieve(url, path.rstrip('\n'))
    print('downloaded: '+val+'_gml.zip')
```
