```python
import pandas as pd
from collections import defaultdict
from collections import Counter
import os
```

## Find most common types per country/region

```python
path_directory = r"D:\Nicola project\Nicola project\mapping\type"
city = []
types = []
for i in os.listdir(path_directory):   # os.listdir(path_directory)
        domain = os.path.abspath(path_directory) 
        region = i.split('.')[0]
        print('region is {}'.format(region))
        city.append(region)
        stat= os.path.join(domain,i) 
        if stat.endswith(".csv"):
            df = pd.read_csv(stat)
            df = df.dropna(subset=['type_source'])
            c = dict(Counter(df.type_source).most_common(50))
            types.append(c)

```

```python
osm =pd.DataFrame()
osm['city'] = city
osm['types'] = types
osm.to_csv('type.csv')
```

## Aggregate

```python
df = pd.read_csv(r'type.csv')
df
```

```python
eval(df['types'][0])
    
```

```python
## agregate

dd = defaultdict(list)

for i in df['types']: # you can list as many input dicts as you want here
    d  = eval(i)
    for key, value in d.items():
        dd[key].append(value)
        
print(dd)
```

```python
## sum up

dd = {i:sum(dd[i]) for i in dd.keys()}
dd
```

```python
## order

data = dict(sorted(dd.items(), key=lambda item: item[1], reverse=True))
data
```

```python
types_data = pd.DataFrame.from_dict(data, orient='index',columns=['sum'])
types_data
```

```python
types_data = types_data.reset_index()
types_data
```

```python
lst1 = []
for i in types_data['index']:
    lst1.append(('house' in i)|('residential' in i)|('apartments' in i)\
               |('bungalow' in i)|('allotment_house' in i)|('dormitory'in i)|('apartment_building' in i)\
               |('home'in i))
types_data['residential'] = lst1
```
